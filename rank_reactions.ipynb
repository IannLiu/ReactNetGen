```python
from NetGenerator.rank_reactions import form_data, rank_reactions, show_results, select_reactions
import NetGenerator.GenNet as GenNet
from NetGenerator.GenNet import str_to_mol
import rdkit
from rdkit import Chem
import pandas as pd
```

```python
# Giving a reactant, generating the products
react_dic, react_dic_reverse = GenNet.read_temp('./templates/rmg_template.txt')
reactant = GenNet.assign_map_num('C1CC[CH]C1=O')
# reactant = GenNet.assign_map_num('[CH2]CCC[C]=O')
# reactant = GenNet.assign_map_num('[CH2]C=C')
display(str_to_mol(reactant))
print(reactant)
outcomes = GenNet.run_one_rank(reactant, react_dic, react_dic_reverse, use_forward_temp=True, use_reverse_temp=True)
for outcome in outcomes:
    mol = str_to_mol(outcome)
    display(mol)
```

```python
# outcomes.append("[C:1]([H:7])([H:8])=[C:2]([H:9])([H:10]).[C:3]([H:11])([H:12])=[C:4]([H:13])([H:14]).[C-:5]#[O+:6]")
# outcomes.append("[C:1]1([H:7])([H:8])[C:2]([H:9])([H:10])[C:3]([H:11])([H:12])[C:4]([H:13])=[C:5]1[O:6][H:14]")
reaction_df = form_data([reactant], outcomes, [1393])
reaction_df.head()
print(reaction_df.size)
```

```python
# path = r'./reactranker/model_results/final_results/qm_rmg_evidential_ranking_penalty'
# path = r'./reactranker/model_results/final_results/qm_rmg_evidential_ranking'
path = r'./reactranker/model_results/final_results/evidential_ranking'
sort_smile_with_map_and_scores, sort_smile = rank_reactions(path, reaction_df, fold_number=10)
```

```python
print(len(sort_smile))
show_results(sort_smile_with_map_and_scores)
```

```python
cut_off = 2
selected, expanded = select_reactions(reaction_batch=sort_smile_with_map_and_scores, cut_off=cut_off)
```
